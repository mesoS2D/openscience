```python
import glob
from PIL import Image, ImageDraw
import time 
import glob
```

```python
days_in = ['2010-02-27']
model_str1 = 'R1'
model_str4 = model_str3 = model_str2 = model_str1
var = 'w'#,'T'] # Only work on single variables at a time
lev_list_in = ['302.00000000','2.96000000','0.00743000','0.00004750']
path_in = "./media/plots/" # "/nobackup/phymku/WP5/WACCM-RR/analysis/maps_2/media/plots/
path_out = "./media/canvases/" # "/nobackup/phymku/WP5/WACCM-RR/analysis/maps_2/media/canvases/

files1 = []
files2 = []
files3 = []
files4 = []

for day in days_in: # In case creating a movie over multiple days
    
    # glob relevant files and append list of files to parent list
    files1.append(sorted(glob.glob(path_in +f'*{model_str1}*{var}*{lev_list_in[0]}*{day}*'))) 
    files2.append(sorted(glob.glob(path_in +f'*{model_str2}*{var}*{lev_list_in[1]}*{day}*')))
    files3.append(sorted(glob.glob(path_in +f'*{model_str3}*{var}*{lev_list_in[2]}*{day}*')))
    files4.append(sorted(glob.glob(path_in +f'*{model_str4}*{var}*{lev_list_in[3]}*{day}*')))

# Flatten lists to have a single list of files, not a list of lists   
files1 = [file for sublist in files1 for file in sublist] 
files2 = [file for sublist in files2 for file in sublist] 
files3 = [file for sublist in files3 for file in sublist] 
files4 = [file for sublist in files4 for file in sublist] 
```

```python
files4
```

```python
tic = time.time()
for i in range(96):
    
    istr1 = files1[i].find("R1") # Get start of filename
    istr2 = files1[i].find(f"_{var}_") + 3 # Identify start of pressure level string so that it can be skipped
    istr3 = files1[i].find("2010") # Identify end of pressure level string so that it can be skipped
    
    fname = files1[i][istr1:istr2] + files1[i][istr3:]
    print(f'Creating canvas for {fname}')
    
    single_snapshot_files = [files1[i], files2[i], files3[i], files4[i]] # Get access to same snapshot for all pressure levels
    
    images = [Image.open(file) for file in single_snapshot_files] # Open all snapshots 
    
    # Get dimension of canvas
    total_width = sum(img.width for img in images)
    height = images[0].height   

    # Create a new white background image
    new_image = Image.new("RGB", (total_width, height), "white")
    
    # Paste images side by side
    x_offset = 0
    for img in images:
        new_image.paste(img, (x_offset, 0))
        x_offset += img.width

    # Save
    new_image.save(f'{path_out}{fname}')
    
    # # Display
    # from IPython.display import display
    # display(new_image)
toc = time.time()
print(f'That whole cell took {toc - tic:2f} seconds')   
```
