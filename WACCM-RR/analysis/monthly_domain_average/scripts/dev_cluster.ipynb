```python
from dask.distributed import Client
from dask_jobqueue import PBSCluster
```

```python
# Configure the PBSCluster with updated parameters
cluster = PBSCluster(
    account='P93300043',
    queue='casper',
    walltime='01:00:00',  # Set the walltime as needed
    cores=72,                         # Total cores for the entire job
    memory='400GB',                   # Memory per job (worker)
    processes=12,                    # Number of Dask worker processes - threads per worker calculated as cores/perocesses
    resource_spec='select=2:ncpus=36:mem=200GB',  # Adjust the resource specification as needed
    job_extra_directives=['-j oe'],  # Updated PBS options
)
cluster.scale(1)
```

```python
client = Client(cluster)
```

```python
# Save the scheduler address to a file
with open("scheduler_address.txt", "w") as f:
    f.write(client.scheduler_info()['address'])
```
